```python
PLAYERS = ['O','X']
SQUARES = 9


def print_board(cells):
    print("---------")
    print('| ' + str(cells[0]) + ' ' + str(cells[1]) + ' ' + str(cells[2]) + ' |')
    print('| ' + str(cells[3]) + ' ' + str(cells[4]) + ' ' + str(cells[5]) + ' |')
    print('| ' + str(cells[6]) + ' ' + str(cells[7]) + ' ' + str(cells[8]) + ' |')
    print("---------")


def wins(cells, player):
    win_state = [
        [cells[0], cells[1], cells[2]],
        [cells[3], cells[4], cells[5]],
        [cells[6], cells[7], cells[8]],
        [cells[0], cells[3], cells[6]],
        [cells[1], cells[4], cells[7]],
        [cells[2], cells[5], cells[8]],
        [cells[0], cells[4], cells[8]],
        [cells[2], cells[4], cells[6]],
    ]
    return [player, player, player] in win_state


def check_board_stat(cells):
    x_count = 0
    o_count = 0
    empties = 0
    for cell in cells:
        if cell == 'X':
            x_count += 1
        elif cell == 'O':
            o_count += 1
        else:
            empties += 1
    X_win = wins(cells, PLAYERS[1])
    O_win = wins(cells, PLAYERS[0])
    if abs(x_count - o_count) >= 2 or (X_win and O_win):
        print("Impossible")
    elif X_win:
        return "X wins"
    elif O_win:
        return "O wins"
    elif empties >= 1:
        return "Game not finished"
    else:
        return "Draw"


def convert_to_cells(move):
    int_move = [int(coord) for coord in move]
    # convert col & row to 0-based board instead of 1-based
    i = 3 - int_move[1]  # 3 - row
    j = int_move[0] - 1  # col - 1
    cells_ind = (i * 3) + j
    return cells_ind


def is_input_num(move):
    for coord in move:
        if not coord.isnumeric():
            return False
        elif int(coord) > 3 or int(coord) < 1:
            return False
    return True


def is_input_valid(move):
    for coord in move:
        if int(coord) > 3 or int(coord) < 1:
            return False
    return True


def is_cell_empty(move, cells):
    ind = convert_to_cells(move)
    if cells[ind] != ' ':
        return False
    return True


def make_turn(cells, player):
    move = input("Enter the coordinates: ").split()
    while not is_input_num(move) or not is_input_valid(move) or not is_cell_empty(move, cells): 
        if not is_input_num(move):
            print("You should enter numbers!")
        elif not is_input_valid(move):
            print("Coordinates should be from 1 to 3!")
        else:
            print("This cell is occupied! Choose another one!")
        move = input("Enter the coordinates: ").split()
    cells[convert_to_cells(move)] = player
    print_board(cells) 
    return cells


def tic_tac_toe():
    cells = [" " for _ in range(SQUARES)]
    print_board(cells)
    moves_used = 0
    while check_board_stat(cells) not in ['X wins', 'O wins', 'Draw'] and moves_used < SQUARES:
        if moves_used % 2 == 0:
            cells = make_turn(cells, PLAYERS[0])
        else:
            cells = make_turn(cells, PLAYERS[1])
        moves_used += 1
    print(check_board_stat(cells))


tic_tac_toe()
```
